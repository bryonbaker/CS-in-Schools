# input
This section is demonstrating using the input() command. The result of input() is stored in a variabled called `name`.

On its own input() does not do much good - we need to work out how to do something with the text we input. We'll show you that next...

```
name = input("Please enter your name and press [RETURN]: ")
```

#Using what we input()
We have stored the result in `name`. Now we will show you how to use it by displaying the text on the screen.

We will also start to use comments in the code to help explain what is going on.

`# A comment is text with a hash as the first character`


```
# Prompt the user to enter a name.
name = input("Please enter your name and press [RETURN] ")

# Display the name on the console
print(name)
```

#Enriching what we input
This section shows you how to enrich displaying the text that was entered with additional text plus colour.


```
# Display a message and wait for the user to enter the name.
name = input("Please enter your name and press [RETURN] ")

# Join the text together with '+' - just like we did for printing in colour.
print("Hello " + name)
```

#Revision: Make it colourful!
Let's try some cool formatting using colour!

```
from csinsc import *    # We need this for colour

# Display a message and wait for the user to enter the name.
name = input("Please enter your name and press [RETURN] ")

print("Hello " + name)

# Now print it out in colour.
print(Colour.green + "Hello " + name + ". This time I am displaying colour.")

# Now print it out in two colours.
print(Colour.green + "Hello " + Colour.blue + name + Colour.green + ". This time I am displaying two colours.")
```

#Variables
Let's give some more examples of using variables.

```
print("Let's put a banana in the lunchbox.")
lunchbox = "banana"
print("I have a " + lunchbox + " in my lunchbox.")
print() # Print a blank line.
print("Let's replace the banana with a sandwich.")
lunchbox = "sandwich"
print("I now have a " + lunchbox + " in my lunchbox.")
```

#Variables: Another example
Here is another use of variables, input(), and print()


```
# Display a message and wait for the user to enter the name.
name = input("Please enter your name and press [RETURN] ")

# Create a new variable with the full message we want to print.
# Join the text together with '+' - just like we did for printing in colour.
message = "Hello " + name

# Print the message
print(message)
```
